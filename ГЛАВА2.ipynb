```python
#2.2.
import matplotlib.pyplot as plt

classes = ['power_line', 'brown_field', 'main_trail']
counts = [1422, 200, 200]

plt.figure(figsize=(8, 5))
bars = plt.bar(classes, counts, color=['#d62728', '#ff7f0e', '#2ca02c'])
plt.ylabel('Количество изображений', fontsize=12)
plt.title('Распределение изображений по классам в датасете CaT', fontsize=14)
plt.ylim(0, 1600)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 20,
             f'{int(height)} ({height/1822*100:.1f}%)',
             ha='center', va='bottom', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Готовые значения (на основе 1822 изображений, разбиение 70/30)
classes = ['power_line', 'brown_field', 'main_trail']
train = [995, 140, 140]   # 70% от 1422, 200, 200
test  = [427, 60, 60]     # 30% от 1422, 200, 200

x = np.arange(len(classes))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, train, width, label='Train', color='#4C72B0', edgecolor='black')
bars2 = ax.bar(x + width/2, test,  width, label='Test',  color='#DD8452', edgecolor='black')

# Подписи значений
for bar in bars1:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10,
            f'{int(bar.get_height())}', ha='center', va='bottom', fontsize=10)
for bar in bars2:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10,
            f'{int(bar.get_height())}', ha='center', va='bottom', fontsize=10)

ax.set_xlabel('Класс', fontsize=12)
ax.set_ylabel('Количество изображений', fontsize=12)
ax.set_title('Распределение обучающей и тестовой выборок в датасете CaT', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(classes)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
